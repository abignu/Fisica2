```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#distribuci칩n

def distribucion(n, m):
  #hay que introducir n y m para reproducir la f칩rmula deseada, ejemplo: x_i+1 = ( n * x_i ) % m
  a = np.zeros(10)
  a[0] = 1
  for i in range(0,a.size-1):
    a[i+1] = (n*a[i]) % m
  
  return a

x = distribucion(2**9,10**3)

z = x/10**3
```

```
#una vez generado z lo usamos para generar el algoritmo
codigo = np.array([0,1,0,0,1,0,0,1,0,1])
codigo_copy = codigo
probabilidades = np.zeros(1000) #array de 100 posiciones
p = 0.1

#bucle while para sacar promedio
j = 0
while j < 1000:
  #ahora lo hago con una distribucion uniforme generada con python
  z_prima = np.random.uniform(0,1,10)

  #bucle para ver cuando cambian
  contador = 0

  for i in range(codigo.size):
    #me fijo si supera la probabilidad de cambio o no
    if z_prima[i] > (1-p):
      contador += 1 #aumento uno pq hubo un cambio
      #si la supera chequeo a ver cual es su valor y lo cambio por el otro
      if codigo[i] == 1:
        codigo[i] = 0
      elif codigo[i] == 0:
        codigo[i] = 1
  
  probabilidades[j]  = contador/codigo.size #guardo la probabilidad
  
  j += 1 #sumo 1 y vuelvo a empezar

#vemos is son iguales
print('Media =', np.mean(probabilidades))
print('Desv est치ndar =', np.std(probabilidades))

plt.bar(range(1000),probabilidades)
plt.xlabel('Simulaciones')
plt.ylabel('Probabilidad obtenida')
plt.show()

cuenta = np.array([i for i in range(1000)])
sns.kdeplot(probabilidades)
plt.xlabel('Probabilidad obtenida')
plt.show()
```
