```
#máquina de Turing que comprueba paridad
import numpy as np
```

```
#declaro las principales herramientas
a = ['b', 'A', 'B', 'X', 'Y', 'delta'] #alfabeto
e = ['Ss', 'Sx', 'Sy', 'Sr'] #estados
```

```
msg = ['A', 'B', 'A', 'delta', 'b', 'b', 'b'] #mensaje

#una vez recibido el mensaje lo analizamos con la máquina de Turing
bool = True
estado = e[0] #asigno por default el estado Ss
estado_siguiente = e[0]

#empiezo en la primera posición
idx = 0 
while bool: 
  
  #si me encuentro un 'b' hago lo siguiente en función del estado:
  if msg[idx] == a[0]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[1]: #Sx
      estado = e[1] #dejo el estado en Sx
      estado_siguiente = e[3] #cambio el estado_siguiente a Sr
      msg[idx] = a[1] #cambio el mensaje a 'A'
      idx = idx - 1 #nos movemos a la izquierda
    elif estado_siguiente == e[2]: #Sy
      estado = e[2] #dejo el estado en Sy
      estado_siguiente = e[3] #cambio el estado_siguiente a Sr
      msg[idx] = a[2] #cambio el mensaje a 'B'
      idx = idx - 1 #nos movemos a la izquierda
    
   
  #si me encuentro un 'A' hago lo siguiente en función del estado:
  if msg[idx] == a[1]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[0]: #Ss
      estado = e[0] #dejo el estado en Ss
      estado_siguiente = e[1] #cambio el estado_siguiente a Sx
      msg[idx] = a[3] #cambio el mensaje a 'X'
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[1]: #Sx
      estado = e[1] #dejo el estado en Sx
      estado_siguiente = e[1] #cambio el estado_siguiente a Sx
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[2]: #Sy
      estado = e[2] #dejo el estado en Sy
      estado_siguiente = e[2] #cambio el estado_siguiente a Sy
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[3]: #Sr
      estado = e[3] #dejo el estado en Sr
      estado_siguiente = e[3] #cambio el estado_siguiente a Sr
      idx = idx - 1 #nos movemos a la izquierda
      
  #si me encuentro un 'B' hago lo siguiente en función del estado:
  if msg[idx] == a[2]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[0]: #Ss
      estado = e[0] #dejo el estado en Ss
      estado_siguiente = e[2] #cambio el estado_siguiente a Sy
      msg[idx] = a[4] #cambio el mensaje a 'Y'
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[1]: #Sx
      estado = e[1] #dejo el estado en Sx
      estado_siguiente = e[1] #cambio el estado_siguiente a Sx
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[2]: #Sy
      estado = e[2] #dejo el estado en Sy
      estado_siguiente = e[2] #cambio el estado_siguiente a Sy
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[3]: #Sr
      estado = e[3] #dejo el estado en Sr
      estado_siguiente = e[3] #cambio el estado_siguiente a Sr
      idx = idx - 1 #nos movemos a la izquierda 
    
  #si me encuentro un 'X' hago lo siguiente en función del estado:
  if msg[idx] == a[3]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[3]: #Sr
      estado = e[3] #dejo el estado en Sr
      estado_siguiente = e[0] #cambio el estado_siguiente a Ss
      msg[idx] = a[1] #cambio el mensaje a 'A'
      idx = idx + 1 #nos movemos a la derecha
  
  #si me encuentro un 'Y' hago lo siguiente en función del estado:
  if msg[idx] == a[4]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[3]: #Sr
      estado = e[3] #dejo el estado en Sr
      estado_siguiente = e[0] #cambio el estado_siguiente a Ss
      msg[idx] = a[2] #cambio el mensaje a 'B'
      idx = idx + 1 #nos movemos a la derecha
  
  #si me encuentro un 'delta' hago lo siguiente en función del estado:
  if msg[idx] == a[5]:
    
    print('{} {} posición {} {}'.format(estado, msg, idx, estado_siguiente))
    
    if estado_siguiente == e[0]: #Ss
      estado = e[0] #dejo el estado en Ss
      estado_siguiente = 'h' #cambio el estado_siguiente a parar
      bool = False #paramos el programa
    elif estado_siguiente == e[1]: #Sx
      estado = e[1] #dejo el estado en Sx
      estado_siguiente = e[1] #cambio el estado_siguiente a Sx
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[2]: #Sy
      estado = e[2] #dejo el estado en Sy
      estado_siguiente = e[2] #cambio el estado_siguiente a Sy
      idx = idx + 1 #nos movemos a la derecha
    elif estado_siguiente == e[3]: #Sr
      estado = e[3] #dejo el estado en Sr
      estado_siguiente = e[3] #cambio el estado_siguiente a Sr
      idx = idx - 1 #nos movemos a la izquierda 
  
print('termine')
```
